Scatter Plot with Different Colors: Task: Create a scatter plot of sepal length against sepal width for the Iris dataset. Use different colors for each species to distinguish them. Customize the plot with appropriate labels and a legend

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = pd.read_csv("D:\progamming\Practice\string\matplotlib\iris.csv")
print(data)

# Create a scatter plot with different colors for each species
sns.set(style="whitegrid")
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=data)

# Customize the plot with labels and a legend
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")

# Show the plot
plt.show()
```

Histogram for Feature Distribution: Task: Generate a histogram to visualize the distribution of petal length for all samples in the Iris dataset. Customize the plot with appropriate labels and adjust the number of bins to display the data effectively.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the Iris dataset from the CSV file
data = pd.read_csv("D:\\progamming\\Practice\\string\\matplotlib\\iris.csv")

# Create a histogram of petal length with customizations
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))  # Set the size of the figure

sns.histplot(data["petal_length"], bins=20, kde=True, color='skyblue')  # Customize the number of bins and add KDE

# Customize the plot with labels and title
plt.title("Distribution of Petal Length in the Iris Dataset")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

Box Plot Comparison: Task: Create a box plot to compare the sepal width distributions for each species in the Iris dataset. Customize the plot by assigning different colors to each species, adding appropriate labels, and adjusting the style of the box plot (e.g., adding outliers, changing whisker styles).



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the Iris dataset from the CSV file
data = pd.read_csv("D:\\progamming\\Practice\\string\\matplotlib\\iris.csv")

# Create a box plot of sepal width with customizations
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))  # Set the size of the figure

# Customize the box plot
sns.boxplot(x="species", y="sepal_width", data=data, palette="Set3", showfliers=True, whis=1.5)

# Customize the plot with labels and title
plt.title("Comparison of Sepal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")

# Show the plot
plt.show()
```

Line Plot with Multiple Lines: Task: Generate a line plot to compare the variations of sepal length and petal length across the samples in the Iris dataset. Each line represents a different species. Customize the plot by adding appropriate labels and a legend.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the Iris dataset from the CSV file
data = pd.read_csv("D:\\progamming\\Practice\\string\\matplotlib\\iris.csv")

# Create a line plot to compare sepal length and petal length for each species
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))  # Set the size of the figure

# Customize the line plot with multiple lines for each species
sns.lineplot(x="sepal_length", y="species", data=data, label="Sepal Length", ci=None)
sns.lineplot(x="petal_length", y="species", data=data, label="Petal Length", ci=None)

# Customize the plot with labels and a legend
plt.title("Comparison of Sepal Length and Petal Length by Species")
plt.xlabel("Length (cm)")
plt.ylabel("Species")
plt.legend(title="Features")

# Show the plot
plt.show()
```

Bar Chart for Categorical Data: Task: Create a bar chart to display the frequency of each species in the Iris dataset. Customize the plot with appropriate labels, a title, and different colors for each species.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the Iris dataset from the CSV file
data = pd.read_csv("D:\\progamming\\Practice\\string\\matplotlib\\iris.csv")

# Count the frequency of each species
species_counts = data['species'].value_counts().reset_index()
species_counts.columns = ['species', 'count']

# Set custom colors for each species
colors = sns.color_palette("Set3")

# Create a bar chart to display the frequency of each species
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))  # Set the size of the figure

# Customize the bar chart
sns.barplot(x='species', y='count', data=species_counts, palette=colors)

# Customize the plot with labels and title
plt.title("Frequency of Iris Species")
plt.xlabel("Species")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```
